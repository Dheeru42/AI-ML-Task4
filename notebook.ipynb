# Importing Libraries

```python
import pandas as pd #useful for loading the dataset
import numpy as np #to perform array
```

# Load Dataset

```python
dataset = pd.read_csv('data.csv')
```

```python
# Check Shape
dataset.shape
```

```python
# Look Dataset
dataset.head(10)
```

```python
dataset['diagnosis'].value_counts()
```

# Mapping Class String Values to Numbers

```python
dataset['diagnosis'] = dataset['diagnosis'].map({'B': 0, 'M': 1}).astype(int)
```

```python
# Check Data after encoding daignosis column
dataset.head(10)
```

# Train Model

```python
# Segregate Dataset into X(Input/IndependentVariable) & Y(Output/DependentVariable)
X = dataset.iloc[:, 2:32].values
y = dataset.iloc[:,1].values
```

```python
# Splitting Dataset into Train & Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) 
X_test = sc.transform(X_test)
```

```python
# import and fit model
from sklearn.linear_model import LogisticRegression 
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
# Check Score at training data
lr.score(X_train,y_train)*100
```

```python
# Check Score at testing data
lr.score(X_test,y_test)*100
```

```python
# Prediction on all testing data
y_pred = lr.predict(X_test) 
```

# Evaluate Model with confusion matrix, precision, recall, ROC-AUC.

```python
# Import metrics
from sklearn.metrics import confusion_matrix,roc_auc_score,roc_curve,precision_score, recall_score
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
# precision, recall
precision = precision_score(y_test,y_pred)*100
recall = recall_score(y_test,y_pred)*100
print("Precision",precision)
print("Recall",recall)
```

```python
# ROC-AUC score
roc_auc = roc_auc_score(y_test, y_pred)*100
print("ROC-AUC Score:", roc_auc)
```

```python
# 4. ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid()
plt.show()
```

# Try different thresholds

```python
# Try different thresholds
thresholds = np.arange(0.1, 1.0, 0.1)

print("Thresh | Precision | Recall")
for thresh in thresholds:
    y_pred = (y_pred >= thresh).astype(int)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    print(f"{thresh:.1f}     | {precision:.2f}      | {recall:.2f} ")
```
